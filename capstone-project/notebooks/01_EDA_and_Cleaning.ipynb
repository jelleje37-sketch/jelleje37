# 1️⃣ Exploratory Data Analysis & Cleaning

## Objective
- Load raw data
- Explore structure and distributions
- Identify and handle missing values
- Prepare clean dataset for analysis

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print('Libraries imported successfully')
```

## Load Data

```python
# Load dataset
df = pd.read_csv('../data/raw/[dataset_name].csv')

# Display basic info
print(f'Dataset shape: {df.shape}')
print(f'\nColumns: {df.columns.tolist()}')
print(f'\nData types:\n{df.dtypes}')
print(f'\nFirst few rows:')
df.head()
```

## Check for Missing Values

```python
# Missing value summary
missing = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': df.isnull().sum(),
    'Missing_Percent': (df.isnull().sum() / len(df) * 100).round(2)
})

missing = missing[missing['Missing_Count'] > 0].sort_values('Missing_Percent', ascending=False)
print('Missing Values Summary:')
print(missing)

# Visualization
plt.figure(figsize=(10, 6))
missing.set_index('Column')['Missing_Percent'].plot(kind='barh', color='coral')
plt.xlabel('Percentage Missing (%)')
plt.title('Missing Values by Column')
plt.tight_layout()
plt.show()
```

## Data Cleaning Steps

```python
# Make a copy for cleaning
df_clean = df.copy()

# Remove duplicates
print(f'Duplicates removed: {df_clean.duplicated().sum()}')
df_clean = df_clean.drop_duplicates()

# Handle missing values (customize based on domain knowledge)
# Option 1: Drop rows with missing values
# df_clean = df_clean.dropna()

# Option 2: Fill with median/mean
# df_clean['numeric_column'] = df_clean['numeric_column'].fillna(df_clean['numeric_column'].median())

# Option 3: Fill with mode
# df_clean['category_column'] = df_clean['category_column'].fillna(df_clean['category_column'].mode()[0])

print(f'\nCleaned dataset shape: {df_clean.shape}')
print(f'Remaining missing values: {df_clean.isnull().sum().sum()}')
```

## Exploratory Analysis

```python
# Descriptive statistics
print('Descriptive Statistics:')
print(df_clean.describe())
```

```python
# Distribution plots for numeric columns
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns

fig, axes = plt.subplots(len(numeric_cols), 1, figsize=(10, 4*len(numeric_cols)))
for idx, col in enumerate(numeric_cols):
    axes[idx].hist(df_clean[col], bins=30, color='skyblue', edgecolor='black')
    axes[idx].set_title(f'Distribution of {col}')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Save Cleaned Data

```python
# Export cleaned dataset
df_clean.to_csv('../data/processed/[cleaned_dataset].csv', index=False)
print('Cleaned dataset saved to data/processed/')
print(f'Final dataset: {df_clean.shape[0]} rows × {df_clean.shape[1]} columns')
```
